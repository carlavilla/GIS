##<b>Changes of property types in Philadelphia</b>

Previously, I looked into the changes in homelessness counts yearly in Philadelphia. Results pointed at a mild postive correlation between the demolition of properties in a given year and the number of reported count of unsheltered homeless inidividuals the following year. Over time, there was also an apparent decrease in the availability of transitional housing and an increase in the availablility and use of emergency housing instead.</br>
For this GIS project, I would like to focus on the changes in property types available to residents (including unhoused folks) in Philadelphia. I would like to visualize the types of building structures available in Philadelphia. Depending on the data I can find, I'd like to start off looking at the density of commercial vs residential spaces, rental vs owned properties, and emergency vs transitional housing. I would also like to map out the areas of demolished properties over the years. </br>
I hypothesize that there might be potential replacement and displacement of transitional and emergency housing in favor of construction that pushes for economic growth, such as commercial and rental properties.
